```python
def read_cell(x, y):
    with open("Averages_empower_final_versiongraphs.csv", 'r') as f:
        reader = csv.reader(f)
        y_count = 0
        for n in reader:
            if y_count == y:
                cell = n[x]
                return cell
            y_count += 1

#print (read_cell(0, 1)) 
```
